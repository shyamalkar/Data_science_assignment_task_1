#1D array operation 

```python
import numpy as np


# Create a 1D array from 1 to 20
arr_1d = np.arange(1, 21)

# a. Calculations
print(f"Sum: {np.sum(arr_1d)}")
print(f"Mean: {np.mean(arr_1d)}")
print(f"Median: {np.median(arr_1d)}")
print(f"Std Dev: {np.std(arr_1d)}")

# b. Indices where elements > 10
indices = np.where(arr_1d > 10)
print(f"Indices > 10: {indices}")
```

#2D Array Operations

```python
# Create a 4x4 array (1-16)
arr_2d = np.arange(1, 17).reshape(4, 4)

# a. Print array
print("Original Array:\n", arr_2d)

# b. Transpose
print("Transpose:\n", arr_2d.T)

# c. Row-wise and column-wise sums
print("Row-wise sums:", np.sum(arr_2d, axis=1))
print("Column-wise sums:", np.sum(arr_2d, axis=0))
```

#Random Arrays & Math

```python
# Create two 3x3 random arrays (1-20)
arr1 = np.random.randint(1, 21, (3, 3))
arr2 = np.random.randint(1, 21, (3, 3))

# a. Element-wise operations
print("Sum:\n", arr1 + arr2)
print("Difference:\n", arr1 - arr2)
print("Product:\n", arr1 * arr2)

# b. Dot product
print("Dot Product:\n", np.dot(arr1, arr2))
```

#Reshaping and slicing 

```python
# Reshape 12-size array to 3x4
arr_reshaped = np.arange(12).reshape(3, 4)

# Slice first two rows and last two columns
slice_arr = arr_reshaped[:2, -2:]
print("Sliced Array:\n", slice_arr)
```

#Assignment 2: Working with pandas. 

```python
import pandas as pd

# 1. Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# a. First five rows
print(df.head())

# b. Summary statistics
print(df[['Age', 'Salary']].describe())

# c. Avg salary in HR
hr_avg = df[df['Department'] == 'HR']['Salary'].mean()
print(f"Average HR Salary: {hr_avg}")

# 2. Add 'Bonus' column (10% of salary)
df['Bonus'] = df['Salary'] * 0.10

# 3. Filter age 25-30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# 4. Group by Department and get avg salary
dept_avg = df.groupby('Department')['Salary'].mean()

# 5. Sort by salary (ascending) and save to CSV
df.sort_values(by='Salary', ascending=True).to_csv('sorted_employees.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# 1. Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y)
plt.title("Growth Trend")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()

# 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color=['blue', 'green', 'red', 'purple'])
plt.title("Student Marks")
plt.show()

# 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = (0.1, 0, 0, 0) # Highlight North America
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title("Revenue Distribution")
plt.show()

# 4. Histogram
random_data = np.random.randint(1, 101, 1000)
plt.hist(random_data, bins=20, edgecolor='black')
plt.title("Frequency Distribution")
plt.show()
```
