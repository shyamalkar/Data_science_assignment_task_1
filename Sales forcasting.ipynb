```python
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = df['Order Date'].dt.year
df['State'] = df['Order Date'].dt.month
```

```python
# Assuming the data is sorted by date
train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:len(df)]
```

**`*Apply andForecasting Model*`**

```python
from sklearn.linear_model import LinearRegression

# Training
X_train = train[['Ship Date', 'State']]
y_train = train['Sales']
model = LinearRegression().fit(X_train, y_train)

# Forecasting
X_test = test[['Ship Date', 'State']]
predictions = model.predict(X_test)
```

***`Evaluate Performance`***

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(test['Sales'], predictions))
r2 = r2_score(test['Sales'], predictions)

print(f"RMSE: {rmse}")
print(f"R² Score: {r2}")
```

```python
import pandas as pd

df = pd.read_csv('stores_sales_forecasting 2.csv', encoding='latin1' )
df
```

***`Data Preprocessing`***

Handling missing value


```python
df.fillna(method='ffill', inplace=True)
```

***`Convert date columns`***

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

**`Split the dataset `**

```python
train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:len(df)]
```

***`Appling a Forecasting Model`***

```python
from statsmodels.tsa.arima.model import ARIMA

# Defining and fitting the model
model = ARIMA(train['Sales'], order=(5,1,0))
model_fit = model.fit()

# Making predictions
predictions = model_fit.forecast(steps=len(test))
```

***`MAE, MSE, RMSE`***

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 1. Calculate Mean Absolute Error (MAE)
# This tells you the average physical error in your sales units
mae = mean_absolute_error(test['Sales'], predictions)

# 2. Calculate Mean Squared Error (MSE) first
mse = mean_squared_error(test['Sales'], predictions)

# 3. Calculate Root Mean Squared Error (RMSE)
# This penalizes larger errors more than MAE does
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
